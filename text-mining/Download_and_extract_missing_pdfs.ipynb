```python
def extract_text_pdf(pdf_file):
    try:
        pdf_text = ''
        with fitz.open(pdf_file) as pdf:
            for page in pdf:
                pdf_text += page.get_text()
    except:
        pdf_text = None

    return pdf_text
```
